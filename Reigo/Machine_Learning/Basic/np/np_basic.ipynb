## 矩阵乘法

```python
#ndarray
a=np.ones((2,2))
b=np.arange(0,4).reshape(2,2)

#矩阵每个元素相乘
c=a*b
d=np.multiply(a,b) #c==d

#矩阵点乘(矩阵乘法)
e=np.dot(a,b)
f=a@b 
g=a.dot(b) #e==f==g
```
